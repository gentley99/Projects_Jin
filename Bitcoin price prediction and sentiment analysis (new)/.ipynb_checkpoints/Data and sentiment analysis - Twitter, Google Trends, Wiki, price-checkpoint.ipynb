```python
# function to get tweets from their acount
def tweet_account(user, number_tweets): 
    tweets = tweepy.Cursor(api.user_timeline, screen_name = user,
                       tweet_mode='extended').items(number_tweets)

    tweet_list = []
    tweet_list_time = []
#     tweet_list_id = []

    for tweet in tweets:
        tweet_list.append(tweet.full_text)
        tweet_list_time.append(tweet.created_at)
#         tweet_list_id.append(tweet.id)
    
    #Create a dataframe to store the tweets with a column called tweets
    df = pd.DataFrame(tweet_list, columns=['tweet'])
    # df
    df['time'] = tweet_list_time
#     df['id'] = tweet_list_id
    df['user'] = user

    return df
```

# Cleaning the twitter data and performing a sentiment analysis

```python
import re
import emoji
```

```python
df = pd.read_csv("all_tweets.csv")
df = df.drop("Unnamed: 0", axis=1)
```

```python
# function to clean text
def cleanTxt(text):
    text = re.sub(r'https?:\/\/\S+', '', text) # remove URLs
    text = re.sub(r'rt|cc', '', text)  # remove RT and cc
    text = re.sub(r'#\S+', '', text)  # remove hashtags
    text = re.sub('\\n', ' ', text) #removes the '\n' string
    text = re.sub(r'@\S+', '', text)  # remove mentions
    text = re.sub('[!?"#\$%&\'()*+,/:@#$%^&*:-]', '', text)  # remove punctuations    
    text = re.sub(r'\s+', ' ', text)  # remove extra whitespace
    text = re.sub(r'\s+$', '', text)  # remove extra whitespace    

    return text

# function to remove emojis
def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
```

```python
# lower casing
df['cleaned_tweet'] = df['tweet'].str.lower() 
# removing emojis
df['cleaned_tweet'] = df['cleaned_tweet'].apply(deEmojify)
# cleaning the text
df['cleaned_tweet'] = df['cleaned_tweet'].apply(cleanTxt)
```

### Stop words

```python
import nltk
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
df['cleaned_tweet_stop'] = df['cleaned_tweet'].apply(lambda x: " ".join(word for word in x.split() if word not in stop_words))
```

```python
pd.Series(" ".join(df['cleaned_tweet_stop']).split()).value_counts()[:10]
```

```python
# getting rid of words that doesn't have much meaning
other_stop_words = ['bitcoin', 'bitcoin.', 'people', 'crypto', 'one', 'amp;', '1', 'would', 'btc', 'woh', 'it.', 'us', 'r.', "it's"]
```

```python
df['cleaned_tweet_stop'] = df['cleaned_tweet_stop'].apply(lambda x: " ".join(word for word in x.split() if word not in other_stop_words))
```

```python
freq_words = pd.Series(" ".join(df['cleaned_tweet_stop']).split()).value_counts()[:30]
```

### Stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])
```

```python
df['cleaned_tweet_stop_stem'] = df['cleaned_tweet_stop'].apply(lambda text: stem_words(text))
```

### Lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
df['cleaned_tweet_stop_lem'] = df['cleaned_tweet_stop'].apply(lambda text: lemmatize_words(text))
```

## Sentiment Analysis

```python
from textblob import Word, TextBlob
```

```python
#Create a function to get the subjectivity
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

# Create a function to get the polarity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity

# Create two new columns
df['Subjectivity'] = df['cleaned_tweet_stop_lem'].apply(getSubjectivity)
df['Polarity'] = df['cleaned_tweet_stop_lem'].apply(getPolarity)
```

```python
#Create a function to compute the negative, neutral and positive analysis
def getAnalysis(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else: 
        return 'Positive'
    
df['Analysis'] = df['Polarity'].apply(getAnalysis)
```

Polarity is float which lies in the range of [-1,1] where 1 means positive statement and -1 means a negative statement.
